**De Jong Function 1 (Sphere Function)**

The de Jong function number 1, often referred to as the Sphere function, is a benchmark problem in optimization. The function is defined as:

$$ f(\mathbf{x}) = \sum_{i=1}^{n} x_i^2 $$

Where:
- $ f(\mathbf{x}) $ is the function value.
- $ x_i $ represents each dimension of the input vector $ \mathbf{x} $.
- $ n $ is the number of dimensions of the input vector $ \mathbf{x} $.

The function's global minimum is 0, which occurs when $ x_i = 0 $ for all $ i $. The Sphere function is convex and unimodal, making it relatively simple compared to other benchmark functions.

---



```python
# Creating a DeJong object with the function number 1.
DJ1 = DeJong(1)
# Printing the function name, global minimum, dimension, range interval, resolution factor, number of bits for dimension,
# and number of bits for individual. It is also plotting the function.
print(
  f'{DJ1.function_name},\nGlobal minimum = 0,\nDimension (genes): {DJ1.dimension},\nRange interval: {DJ1.range},\nResolution factor: {DJ1.resolution_factor},\nNumber of bits for dimension: {DJ1.num_bits},\nNumber of bits for individual: {DJ1.num_bits * DJ1.dimension}.')
DJ1.plot()
```

##De Jong Function 2 (Rosenbrock's Valley)

The de Jong function number 2, commonly known as the Rosenbrock's Valley or the Rosenbrock function, is another benchmark problem in optimization. The function is defined as:

$$ f(\mathbf{x}) = \sum_{i=1}^{n-1} \left[ 100(x_{i+1} - x_i^2)^2 + (1 - x_i)^2 \right] $$

Where:
-  $f(\mathbf{x})$ is the function value.
-  $x_i$ represents each dimension of the input vector $\mathbf{x}$.
- $n$ is the number of dimensions of the input vector $\mathbf{x}$.

The function has a global minimum of 0, which occurs at $x_i = 1$  for all $i$. The Rosenbrock function is non-convex and has a narrow, curved valley. The bottom of this valley is flat, and it is easy to find the valley but challenging to converge to the global minimum.

---



```python
# Creating a DeJong object with the function number 2.
DJ2 = DeJong(2)
print(
  f'{DJ2.function_name} or Rosenbrock’s valley or banana function,\nGlobal minimum = 0,\nDimension (genes): {DJ2.dimension},\nRange interval: {DJ2.range},\nResolution factor: {DJ2.resolution_factor},\nNumber of bits for dimension: {DJ2.num_bits},\nNumber of bits for individual: {DJ2.num_bits * DJ2.dimension}.')
DJ2.plot()
```

**De Jong Function 3 (Step Function)**

The de Jong function number 3, commonly referred to as the Step function, is another benchmark problem in optimization. The function is defined as:

$$ f(\mathbf{x}) = \sum_{i=1}^{n} \lfloor x_i \rfloor $$

Where:
- $ f(\mathbf{x}) $ is the function value.
- $ x_i $ represents each dimension of the input vector $ \mathbf{x} $.
- $ n $ is the number of dimensions of the input vector $ \mathbf{x} $.
- $ \lfloor \cdot \rfloor $ denotes the floor function, which rounds its argument down to the nearest whole number.

The function's global minimum is when the $\mathbf{x}$ is equal to (-5.12, -5.12, -5.12, ...) and $f(\mathbf{x}) = 0$. The Step function is non-continuous and non-differentiable, making it a challenging problem for some optimization algorithms.

---

Where:
- $x_1$  and $x_2$ are the dimensions of the input vector $\mathbf{x}$.
- $a$ is a 2x25 matrix with specified values.

This function contains 25 local minima. It's a 2-dimensional function and is often used to test the ability of optimization algorithms to escape local minima.

> # Gray Code (Gray Encoding)
>
> Gray code is a binary numeral system where two successive numbers differ in only one bit. This makes it particularly useful in scenarios where it's beneficial to avoid sudden changes in multiple bits, such as analog-to-digital converters.
>
> ## Properties of Gray Code:
>
> 1. **Uniqueness:** Every binary number has a unique Gray code equivalent.
> 2. **Single-bit change:** Only one bit changes at a time when moving from one value to the next.
>
> ## How to Convert Binary to Gray Code:
>
> To convert a binary number to Gray code:
>
> 1. The MSB (most significant bit) of the Gray code is the same as the original number.
> 2. The subsequent bits of the Gray code can be obtained by XOR-ing the current bit with the next significant bit.
>
> For example, to convert the binary number `1101` to Gray code:
>
> 1. The MSB is `1`.
> 2. XOR the first and second bits: `1 ⊕ 1 = 0`.
> 3. XOR the second and third bits: `1 ⊕ 0 = 1`.
> 4. XOR the third and fourth bits: `0 ⊕ 1 = 1`.
>
> Thus, the Gray code representation is `1011`.
>
> ## How to Convert Gray Code back to Binary:
>
> 1. The MSB of the binary code is the same as the Gray code.
> 2. To get the subsequent binary bits, XOR the current Gray code bit with the previous binary bit.
>
> For example, to convert the Gray code `1011` back to binary:
>
> 1. The MSB is `1`.
> 2. XOR the MSB with the next Gray code bit: `1 ⊕ 0 = 1`.
> 3. XOR the previous result with the next Gray code bit: `1 ⊕ 1 = 0`.
> 4. XOR the previous result with the next Gray code bit: `0 ⊕ 1 = 1`.
>
> Thus, the binary representation is `1101`.

### Genetic Algorithm Class

```python
# It's a class that implements a genetic algorithm for the De Jong function
class GeneticAlgorithmForDJ:

  def __init__(self, dj_instance, selection_rule, elitism=False, population_array=None, population_size=20,
               mutation_rate=0.01, iterations=600):
    """
    This function initializes the genetic algorithm with the given parameters.

    :param dj_instance:      This is the instance of the De Jong model that you want to optimise.
    :param selection_rule:   The selection rule to be used. Can be either 'rank' or 'roulette'.
    :param elitism:          True or False. If True, the best individual in the population will be carried over
                             to the next generation. Defaults to False (optional), in this case only the population
                             is replaced by the new generation.
    :param population_array: Optional parameter, can contain an initial array with the population. It is a list of
                             lists, where each inner list correspond to an individual of the population.
    :param population_size:  The number of individuals in the population, defaults to 20 (optional).
    :param mutation_rate:    The probability of a gene mutating, defaults 0.01 (optional).
    :param iterations:       The number of iterations to run the algorithm for, defaults to 600 (optional).
    """

    self.dj_instance = dj_instance
    self.dimension = self.dj_instance.dimension
    self.dimension_bits = self.dj_instance.num_bits
    self.selection_rule = selection_rule
    self.population_size = population_size
    self.population_array = population_array
    self.mutation_rate = mutation_rate
    self.iterations = iterations
    self.elitism = elitism

    # Encoding functions
    # Vectorized encoding functions
    self.encoding_function = np.vectorize(self.dj_instance.gray_encode)
    self.decoding_function = np.vectorize(self.dj_instance.gray_decode)

    # Statistics variables
    
    # hold best solution for ith iteration
    # then the best solution for i+1ith iteration
    # will be overwritten on it
    self.best_solution=None
    # best solution decoded
    self.best_solution_decoded=None
    # best value objective function
    self.best_value=None

    # mean value and standard deviation of objective function for ith iteration
    self.mean_value=None
    self.std=None

    # best fitness ith iteration
    self.best_fitness=None
    
    # list of best values across generations
    self.best_value_list=[]
    # list of best solutions across generations
    self.best_solution_list=[]
    # list of generation mean and std for each generation
    self.mean_value_list=[]
    self.std_list=[]
    # best objective value ever achieved
    self.best_value_ever=None
      
    # Additional variables
    # variable to hold population fitness of current generation
    self.population_fitness=None
    # variable to hold probability of current generation
    self.probability=None
    # parents of current generation
    self.parents=None
    # children of current generation
    self.children=None
    # mutated population of current generation
    self.mutated_population=None
    # mutaded+ population_array
    self.combined_population=None

    
    

  def solve(self):
    """
    This function is the main of the class `GeneticAlgorithm`. It performs the main loop of the algorithm,
    and it calls the other functions of the class.
    """
    # Initialise population
    # ...
    self.population_array=self.initialise_population()

    # Evaluate population
    # ...
    self.population_fitness=self.compute_fitness(self.population_array)

    # compute stats for first generation
    self.update_statistics()
      
    for iteration in range(self.iterations):

  # Select parents
  # ...
        self.selection()
  # Generates new individuals using reproduction rules
  # ...
        self.crossover()
        self.mutation()
  # Form a new population
  # ...
        self.reconstruction()
  # Evaluate population
  # ...
        # print('#iteration: ',iteration)

    # print("New Population: ", self.population_array)
        
        self.update_statistics()

        # compute fitness of new population, this information will be used
        # in the next iteration of self.selection()
        self.population_fitness=self.compute_fitness(self.population_array)


    # get index of best value achieved over all generations
    best_index=np.argmin(self.best_value_list)
    # best objective value ever achieved
    self.best_value_ever=self.best_value_list[best_index]
    # corresponding best solution ever
    self.best_solution_ever=self.best_solution_list[best_index]
    # accumulate minimum of best_value_list in order to keep track of its improvement
    self.best_value_list_ever=np.minimum.accumulate(self.best_value_list)

    # iteration at which the best value is achieved
    self.best_iteration=best_index.copy()
        # print('####################')
  def initialise_population(self):
    """
    This function generates a random population of shape `population_size` and `dimension` with values between
    `range[0]` and `range[1]`.
    :return: The population encoded using gray code.
    """
    # initialize randomly population
    initial_population=np.random.uniform(self.dj_instance.range[0], self.dj_instance.range[1], (self.population_size, self.dimension))
    # gray encode the generated population and return it
    gray_initial_population=self.encoding_function(initial_population)
    return gray_initial_population

  def compute_fitness(self, population):
    """
    :return:
    """
    # compute objective
    objective_function = self.dj_instance.evaluate(population, gray_=True)
    # compute fitness
    # my fitness is 1/objective_function
    # In this case the inverse of the objective function is safe because
    # because all DeJong1, DeJong2, DeJong3 and DeJong5 are positive functions
    fitness_population =  1/(1+objective_function)
      
    return fitness_population

  def rank_selection(self):
    """
    This function compute the individual probability of selection using rank selection rule.
    :return: The probability of each individual being selected for the next generation.
    """
    # get size of population
    n = self.population_size

    # argsort along self.population_fitness
    ranking_idx = np.argsort(self.population_fitness)

    # sum of all ranks from 1 to n
    den=n*(n+1)/2

    # create ranking list that assign rank_i to the ith position
    ranking=np.empty_like(ranking_idx)
    ranking[ranking_idx] = np.arange(len(ranking))

    # add 1 to sorted indexes so that ranking goes from 1 to n
    # and compute probability
    probability=(ranking+1)/den
    # print(probability, den)

    return probability

  def roulette_selection(self):
    """
    This function compute the individual probability of selection using roulette selection rule.
    :return: The probability of each individual being selected for the next generation.
    """
    # probabilty with roulette rule
    probability=self.population_fitness/np.sum(self.population_fitness)

    return probability

  def selection(self):
    """
    This function is used to select the parents for the next generation according to a selection rule.
    """
    # Computing the probability of each individual to be selected as parent.
    # ...

    # Selecting the parents for the next generation using probabilities.
    # ...
    
    # select the desired method to use for assigning probabilities
    if self.selection_rule=='rank':
        probability=self.rank_selection()
    elif self.selection_rule=='roulette':
        probability=self.roulette_selection()

    # randomly choose the parents, with replacement, from the population pool
    selected_parents=np.random.choice(np.arange(self.population_size), size=self.population_size,
                                      p=probability)
    # set the selected parents
    self.parents=self.population_array[selected_parents]
    
  def crossover(self):
    """
    This function perform a single-point crossover on each pair of parents, selected by pairing the first half of
    the parents with the second half of them.
    """
    children=[]
    
    # floor the population size to get max number of possible pairs considered the settings
    number_pairs=int(np.floor(self.population_size/2))

    # divide array of parents into two equally long arrays
    # of parents
    parent1=self.parents[:number_pairs]
    parent2=self.parents[number_pairs:]

    # for each pair generate two children
    for i in range(number_pairs):
        child1=[]
        child2=[]

        # for each dimension generate chromosomes of childrens
        for j in range(self.dimension):
            # print(parent1)
            # print(parent2)
            chromosome1=parent1[i][j]
            chromosome2=parent2[i][j]
            # sample an int in interval [1,M) where M is the number of bits for each single
            # dimension, in the Dejong functions all dimensions are represnted by the same number
            # of bits
            crossover_position=np.random.randint(1,self.dimension_bits)
            # 1 point crossover to generate chromosomes of child1 and child2
            child1+=[chromosome1[:crossover_position]+chromosome2[crossover_position:]]
            child2+=[chromosome2[:crossover_position]+chromosome1[crossover_position:]]

        children+=[child1]
        children+=[child2]
    # set the generated children
    self.children=np.array(children.copy())

    

  def mutation(self):
    """
    This function, decide if mutate or not each bit of each offspring in each child, by flip it, with a mutation
     rate, otherwise keep it the same.
    """
    # start from the pool of children
    self.mutated_population=self.children.copy()

    # for each children
    for child_idx, child in enumerate(self.children):
        # for each of its chromosomes
        for chromosome_idx, chromosome in enumerate(child):
            muted=''
            # for each bit of a chromosome
            for bit in chromosome:
                # with probability equal to mutation_rates
                # we convert 1 bit to 0 bit and 0 bit to 1 bit
                if np.random.uniform()<self.mutation_rate:
                    muted+=str(1-int(bit))
                # otherwise keep the bit as it is
                else:
                    muted+=bit
        
        self.mutated_population[child_idx][chromosome_idx]=muted
        
    

  def elitism_selection(self):
    """
    Generating a new population as combination of the old one and the new one.
    """
    # combine the population array with the mutated population
    self.combined_population=np.concatenate([self.population_array, self.mutated_population])
    
    # compute fintess of combined population
    combined_population_fitness=self.compute_fitness(self.combined_population)
    # to get the best indices in terms of fitness, reverse the result of argsort
    # this is because the output of argsort is in ascending order in terms of fitness
    best_idx=np.argsort(combined_population_fitness)[::-1][:self.population_size]

    return self.combined_population[best_idx]

  def reconstruction(self):
    """
    If elitism is enabled, this function replace the population with the 20 best individuals.
    """
    # if elitism is true
    if self.elitism:
        # create the  new population
        # by calling elistism_selection()
        self.population_array=self.elitism_selection()

    else:
        # else the new population_array is the mutated_population
        self.population_array=self.mutated_population

  def update_statistics(self):
    """
    This function can be used to compute statistics on the objective value.
    """
    # get mean of objective value for each generation of population
    self.mean_value=np.average(self.dj_instance.evaluate(self.population_array, gray_=True))
    self.mean_value_list+=[self.mean_value]

    # get std of objective value for each generation of population
    self.std=np.std(self.dj_instance.evaluate(self.population_array, gray_=True))
    self.std_list+=[self.std]
    
    # print(self.mean_value)

    # best fitness across generation
    self.best_fitness=np.max(self.compute_fitness(self.population_array))

    # best solution in the generation
    self.best_solution=self.population_array[np.argmax(self.compute_fitness(self.population_array))]
    
    # best solution in the generation decoded
    self.best_solution_decoded=self.decoding_function(self.best_solution)
      
    # list of best solution for each generation of history
    self.best_solution_list+=[self.best_solution]

    # objective value of best solution across the generation
    self.best_value=self.dj_instance.evaluate(np.array([self.best_solution]), gray_=True)
    
    # list of objective value for each best solution of every generation in the history
    self.best_value_list+=[self.best_value]
    
    

    # print('Best objective function value: ', self.best_value)
    # print('Best solution encoded: ', self.best_solution)
    # print('Best solution decoded: ', self.best_solution_decoded)
    # print('Best fitness: ', self.best_fitness)
```
